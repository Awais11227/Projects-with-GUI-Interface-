```python
import tkinter as tk
from tkinter import messagebox, scrolledtext

class Email:                                                                                            # Define the Email class
    def __init__(self, sender, recipient, subject, body):                                               # Constructor
        self.sender = sender                                                                            # self is an object here 
        self.recipient = recipient
        self.subject = subject
        self.body = body

    def __str__(self):                                                                                  # define a method
        return f"From: {self.sender}\nTo: {self.recipient}\nSubject: {self.subject}\nBody: {self.body}\n"

class EmailClient:                                                                                     # Define the EmailClient class                                                                                    
    def __init__(self):                                                                                # construtor -> super power
        self.emails = []

    def send_email(self, email):                                                                       # Method define
        self.emails.append(email)

    def get_emails(self):                                                                              # Method
        if not self.emails:
            return "No emails found."
        else:
            return "\n\n".join([f"Email {idx}:\n{email}" for idx, email in enumerate(self.emails, 1)])


def send_email():                                                                                   # Function to sending emails
    # Access input 
    sender = entries["sender"].get().strip()
    recipient = entries["recipient"].get().strip()
    subject = entries["subject"].get().strip()
    body = body_text.get("1.0", tk.END).strip()

    # Valid inputs
    if not sender or not recipient or not subject or not body:
        messagebox.showerror("Error", "All fields are required.")
        return

    # Create and send the email
    email = Email(sender, recipient, subject, body)
    client.send_email(email)
    messagebox.showinfo("Success", "Email sent successfully!")

    # Clear input 
    entries["sender"].delete(0, tk.END)
    entries["recipient"].delete(0, tk.END)
    entries["subject"].delete(0, tk.END)
    body_text.delete("1.0", tk.END)

def view_emails():                                                                              # Function to view sent emails
    emails = client.get_emails()
    if emails == "No emails found.":
        messagebox.showinfo("Sent Emails", emails)
    else:
        #  display sent emails
        emails_window = tk.Toplevel(root)
        emails_window.title("Sent Emails")
        emails_window.geometry("600x400")
        emails_window.configure(bg="#f4f4f9")

        emails_text = scrolledtext.ScrolledText(emails_window, wrap=tk.WORD, font=("Arial", 12), bg="#ffffff", fg="#333333")
        emails_text.pack(expand=True, fill=tk.BOTH, padx=10, pady=10)
        emails_text.insert(tk.END, emails)
        emails_text.config(state=tk.DISABLED)

def exit_application():                                                                 #  exit the application
    root.destroy()                                                                      #  closes the application

# Initialize the application
root = tk.Tk()
root.title("ðŸ“§ Email Client by Awais Manzoor")
root.geometry("550x700")
root.configure(bg="#f4f4f9")

# Create an instance 
client = EmailClient()

# Header Label
header_label = tk.Label(root, text="ðŸ“§ Email Client", font=("Arial", 20, "bold"), bg="#f4f4f9", fg="#333333")
header_label.pack(pady=15)

# Input Fields
input_frame = tk.Frame(root, bg="#f4f4f9")
input_frame.pack(padx=20, pady=10, fill=tk.X)

fields = [
    ("Sender:", "sender"),
    ("Recipient:", "recipient"),
    ("Subject:", "subject"),
]

entries = {}

for i, (label_text, field_name) in enumerate(fields):
    tk.Label(input_frame, text=label_text, font=("Arial", 12), bg="#f4f4f9", fg="#333333").grid(row=i, column=0, sticky="w", pady=5)
    entry = tk.Entry(input_frame, font=("Arial", 12), width=40)
    entry.grid(row=i, column=1, pady=5, padx=10)
    entries[field_name] = entry

# Body 
tk.Label(input_frame, text="Body:", font=("Arial", 12), bg="#f4f4f9", fg="#333333").grid(row=len(fields), column=0, sticky="nw", pady=5)
body_text = tk.Text(input_frame, font=("Arial", 12), height=10, width=40, bg="#ffffff", fg="#333333")
body_text.grid(row=len(fields), column=1, pady=5, padx=10)

# Buttons 
buttons_frame = tk.Frame(root, bg="#f4f4f9")
buttons_frame.pack(pady=20)

send_button = tk.Button(buttons_frame, text="Send Email", font=("Arial", 12), bg="#2ecc71", fg="white", command=send_email)
send_button.grid(row=0, column=0, padx=10)

view_button = tk.Button(buttons_frame, text="View Sent Emails", font=("Arial", 12), bg="#3498db", fg="white", command=view_emails)
view_button.grid(row=0, column=1, padx=10)

exit_button = tk.Button(buttons_frame, text="Exit", font=("Arial", 12), bg="#e74c3c", fg="white", command=exit_application)
exit_button.grid(row=0, column=2, padx=10)

# Run  application
root.mainloop()
```
